```python
import numpy as np
import matplotlib.pyplot as plt

rng = np.random.default_rng()
n = 5

X = np.arange(1:n+1)
X = np.insert(X, 0, 0)
# 2(1) - 1 = 1, 2(0)-1 = -1
samples = 2 * rng.binomial(n = 1, p = 0.5, size = n) - 1
Y = np.cumsum(samples)
Y = np.insert(Y, 0, 0)
mean = np.mean(Y)
var = np.var(Y)
se = np.sqrt(var)

max = np.max(Y)
min = np.min(Y)

plt.plot(X, Y)
plt.scatter(X, Y, color = 'r')

plt.axhline(y=0, color='black')

plt.show()
```
